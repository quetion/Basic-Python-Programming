# 練習題
1. 請畫出下列函數的圖形
$$ f(x) = sin(x) $$
2. 請畫出下列函數的圖形
$$ f(x) = cos(x) $$
3. 請畫出下列函數的圖形
$$ f(x) = sin(x) + cos(x) $$
4. 請畫出下列函數的圖形
$$ f_1(x) = sin(x) $$
$$ f_2(x) = sin(2x) $$
$$ f_3(x) = sin(3x) $$

```python
# 第一題解答
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y)
```

```python
# 第二題解答
x = np.linspace(0,10,100)
y = np.cos(x)
plt.plot(x,y)
```

```python
# 第三題解答
x = np.linspace(0,10,100)
y = np.sin(x) + np.cos(x)
plt.plot(x,y)
```

```python
# 第四題解答
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = np.sin(3*x)
a = plt.plot(x,y1) 
b = plt.plot(x,y2) 
c = plt.plot(x,y3)
plt.legend(['sin(x)','sin(2x)','sin(3x)'])
```
